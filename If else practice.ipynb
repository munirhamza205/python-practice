```python
# elif statement
a = 33
b = 33
if b > a:
  print("b is grater than a")
elif a == b:
    print("a and b are equal")
```

```python
# else statement with elif
a = 200
b = 33
if b > a:
  print("b is grater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is grater than b")


```

```python
# else statement
a = 200
b = 33
if b > a:
  print("b is grater than a")
else:
  print("b is not grater than a")
```

```python
# short hand if
a = 200
b = 33
if a > b: print("a is grater than b")
```

```python
# shorthand if..... else

a = 2
b = 330
print("A") if a > b else print("B")
```

```python
# if else with three conditions
a = 330
b = 330
print("A") if a > b else print("=") if a == b else print("B")
```

```python
# doubble if statement
a = 330
b = 330
if a > b:
  print("A")
if a == b:
  print("B")
```

```python
# and operation
a = 200
b = 33
c = 500
if a > b and c > a:
  print("both conditions are true")
```

```python
# or operation
a = 200
b = 33
c = 500
if a > b or a > c:
  print("at least one condition is true")
```

```python
# not operation
a = 33
b = 200
if not a > b:
  print("a is not grater than b")
```

```python
# nested if
x = 41
if x > 10:
  print("Above ten,")
  if x > 20:
    print("and also above 20!")
else:
  print("but not above 20.")
```

```python
# how to remove error
# use pass after if
a = 33
b = 200
if b > a:
  pass
```

```python
NUM1 = float(input("Enter the first number: "))
OPERATOR = input("Enter the operator (+, -, *, /): ")
NUM2 = int(input("Enter the second number: "))

if OPERATOR == '+':
  RESULT = NUM1 + NUM2
elif OPERATOR == '-':
  RESULT = NUM1 - NUM2
elif OPERATOR == '*':
  RESULT = NUM1 * NUM2
elif OPERATOR == '/':
  RESULT = NUM1 / NUM2
else:
  RESULT = "Invalid operator"

print(RESULT)
```
