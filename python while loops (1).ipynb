```python
# the while loops
i = 1
while i <= 6:
  print(i)
  i += 1
print("done")
```

```python
# the break statement
i = 1
while i <= 6:
  print(i)
  if (i == 3):
    break
  i += 1
```

```python
# the continue statement
i = 0
while i < 6:
  i += 1
  if (i == 3):
    continue
  print(i)
  # note that 3 is missing in the result
```

```python
# if statement
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```

```python
# the break statement
numbers = [1, 2, 3, 4, 5]

for number in numbers:
    if number == 3:
        break  # Exit the loop when number is 3
    print(number)
```

```python
# Summing numbers from 1 to 5
total = 0
for i in range(1, 6):
    total += i  # Add i to total for each iteration

print(total)  # Outputs 15
```

```python
#The answer comes to 15 because the code sums the numbers from 1 to 5 through a series of additions. Here’s a step-by-step breakdown of what happens during each iteration of the loop:

#Initialization:

#total starts at 0.
#Iterations:

#Iteration 1: i = 1

#total = 0 + 1 → total = 1
#Iteration 2: i = 2
#total = 1 + 2 → total = 3
#Iteration 3: i = 3
#total = 3 + 3 → total = 6
#Iteration 4: i = 4
#total = 6 + 4 → total = 10
#Iteration 5: i = 5
#total = 10 + 5 → total = 15
```
